```python
import pandas as pd
import os
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, precision_score, recall_score, accuracy_score, confusion_matrix
```

```python
from google.colab import drive
drive.mount('/content/drive')
#load data for PC(sometimes need to use "/")
df = pd.read_csv('/content/STANDARD_DPx1b.csv')
x = pd.concat([df['a'], df['b'], df['c'], df['d'], df['e'], df['f'], df['g'], df['h'], df['i'], df['j'], df['k']
               , df['l']], axis=1)
y = df.drop(['gender', 'age', 'id', 's', 'blood', 'a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l'],
            axis=1).values.ravel()

le = LabelEncoder()
y = le.fit_transform(y)
print(x)
print(y)
```

```python
#12個變數SVM
models = [SVC(gamma='auto', kernel="linear", probability=True),
          SVC(gamma='auto', kernel="poly", probability=True),
          SVC(gamma='auto', kernel="rbf", probability=True),
]

for i, model in enumerate(models, 1):
    print(f"模型 {i}:")
    scores = cross_val_score(model, x, y, cv=10, scoring='accuracy')
    print(f"交叉驗證準確率: {scores}")
    print(f"平均準確率: {round(scores.mean(), 4)}")
    X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2023)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(f"測試準確率: {round(accuracy_score(y_test, y_pred), 4)}")
    print(f"測試F1_score: {round(f1_score(y_test, y_pred, average='weighted'), 4)}")
    print(f"測試Precision: {round(precision_score(y_test, y_pred, average='weighted'), 4)}")
    print(f"測試Recall: {round(recall_score(y_test, y_pred, average='weighted'), 4)}")
    conf_matrix = confusion_matrix(y_test, y_pred)
    confusion_df = pd.DataFrame(conf_matrix,
                                index=['觀光工廠', '都市之旅', '知性之旅', '自然景觀', '休閒娛樂', '文化巡禮'],
                                columns=['觀光工廠', '都市之旅', '知性之旅', '自然景觀', '休閒娛樂', '文化巡禮'])
    print("測試多類别混淆矩陣：")
    print(confusion_df)
    print("\n")
```

```python
#6個變數
x1=pd.concat([df['f'],df['i'],df['j'],df['k'],df['l']],axis=1)
models = [SVC(gamma='auto', kernel="linear", probability=True),
          SVC(gamma='auto', kernel="poly", probability=True),
          SVC(gamma='auto', kernel="rbf", probability=True),
]

for i, model in enumerate(models, 4):
    print(f"模型 {i}:")
    scores = cross_val_score(model, x1, y, cv=10, scoring='accuracy')
    print(f"交叉驗證準確率: {scores}")
    print(f"平均準確率: {round(scores.mean(), 4)}")

    # 因为交叉验证已经在整个数据集上评估了模型，此处可以将数据集划分为训练集和测试集
    X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2023)

    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)
    print(f"測試準確率: {round(accuracy_score(y_test, y_pred), 4)}")
    print(f"測試F1_score: {round(f1_score(y_test, y_pred, average='weighted'), 4)}")
    print(f"測試Precision: {round(precision_score(y_test, y_pred, average='weighted'), 4)}")
    print(f"測試Recall: {round(recall_score(y_test, y_pred, average='weighted'), 4)}")

    conf_matrix = confusion_matrix(y_test, y_pred)
    confusion_df = pd.DataFrame(conf_matrix,
                                index=['觀光工廠', '都市之旅', '知性之旅', '自然景觀', '休閒娛樂', '文化巡禮'],
                                columns=['觀光工廠', '都市之旅', '知性之旅', '自然景觀', '休閒娛樂', '文化巡禮'])
    print("測試多類别混淆矩陣：")
    print(confusion_df)
    print("\n")
```

```python
#12個變數，解決資料不平衡
from imblearn.over_sampling import RandomOverSampler, SMOTE
from imblearn.under_sampling import RandomUnderSampler
kernels = ['linear', 'poly', 'rbf']
samplings = [RandomUnderSampler(random_state=21), RandomOverSampler(random_state=21), SMOTE(random_state=21)]

model_number = 1
for kernel in kernels:
    for sampling in samplings:
        x_resampled, y_resampled = sampling.fit_resample(x, y)
        X_train, X_test, y_train, y_test = train_test_split(x_resampled, y_resampled, test_size=0.2, random_state=2023)

        model = SVC(gamma='auto', kernel=kernel, probability=True)
        model.fit(X_train, y_train)

        scores = cross_val_score(model, x_resampled, y_resampled, cv=10, scoring='accuracy')
        print(f"核函式: {kernel}, 平衡方法: {type(sampling).__name__}")
        print(f"交叉驗證準確率: {scores}")
        print(f"平均準確率: {round(scores.mean(), 4)}")

        y_pred_test = model.predict(X_test)
        print(f"模型 {model_number} 測試準確率=", round(accuracy_score(y_test, y_pred_test), 4))
        print(f"模型 {model_number} 測試F1_score=", round(f1_score(y_test, y_pred_test, average='weighted'), 4))
        print(f"模型 {model_number} 測試Precision:", round(precision_score(y_test, y_pred_test, average='weighted'), 4))
        print(f"模型 {model_number} 測試Recall:", round(recall_score(y_test, y_pred_test, average='weighted'), 4))
        conf_matrix = confusion_matrix(y_test, y_pred_test)
        confusion_df = pd.DataFrame(conf_matrix, index=['觀光工廠', '都市之旅', '知性之旅', '自然景觀', '休閒娛樂', '文化巡禮'],
                                    columns=['觀光工廠', '都市之旅', '知性之旅', '自然景觀', '休閒娛樂', '文化巡禮'])
        print(f"模型 {model_number} 多類别混淆矩陣：")
        print(confusion_df)
        print("-----------------------------------------------------")
        model_number += 1
```
